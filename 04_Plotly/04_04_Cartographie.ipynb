# Cartographie

Recupérons un jeu de données disponible dans la librairie Ploly

```python
#Dataset de la disponibilité des voitures de partage autour de Montréal
df = px.data.carshare()
df.head()
```

```python
fig = px.scatter_mapbox(df, lat="centroid_lat", lon="centroid_lon", color="peak_hour", size="car_hours", zoom=10, mapbox_style="carto-positron")
fig.show()
```
