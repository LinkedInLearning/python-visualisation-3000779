## Import des librairies

```python
#indique à matplotlib d’afficher les graphiques “en ligne”
%matplotlib inline

# Importation des librairies pour la manipulation de données
import numpy as np
import pandas as pd

# Importation des librairies de visualisations
from matplotlib import pyplot as plt
import seaborn as sns 

import warnings
warnings.filterwarnings("ignore")
```

## Création des données

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)
```

# Utiliser des paramètres pour changer l'aspect des courbes

## Changer la couleur

```python
plt.plot(x, y, color="green")
plt.show()
```

## Changer le type de ligne

![](https://matplotlib.org/stable/_images/sphx_glr_linestyles_001.png)

```python
plt.plot(x, y, color="blue", linestyle="--")
plt.show()
```

```python
plt.plot(x, y, color="blue", linestyle="dotted")
plt.show()
```

```python
plt.plot(x, y, color="blue", linestyle=(0, (3, 10, 1, 10, 1, 10)))
plt.show()
```
