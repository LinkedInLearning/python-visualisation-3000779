## Premier graphique

```python
# Tracons un premier graphique

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots() #Permet la création de la figure et du Axes pour la fenêtre graphique
ax.plot(x, y)
plt.show()
```
