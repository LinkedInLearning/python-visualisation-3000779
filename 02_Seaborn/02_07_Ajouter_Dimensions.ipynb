## Import des librairies

```python
#indique à matplotlib d’afficher les graphiques “en ligne”
%matplotlib inline

# Importation des librairies pour la manipulation de données
import numpy as np
import pandas as pd

# Importation de la librairie de visualisation
import seaborn as sns 

import warnings
warnings.filterwarnings("ignore")
```

## Import des données

```python
CheminFichier = r'Data\Dataset_Employes.xlsx'
Dataset_Employes = pd.read_excel(CheminFichier)
Dataset_Employes.head()
```

# Ajouter des dimensions au nuage de points

## Trois dimensions

```python
sns.relplot(x = 'Age' , y = 'Niveau de satisfaction', hue = 'Type Contrat', data = Dataset_Employes)
```

## Quatre dimensions

```python
sns.relplot(x = 'Age' , y = 'Niveau de satisfaction', hue = 'Type Contrat', size = 'Salaire base mensuel', 
            data = Dataset_Employes)
```

## Cinq dimensions

```python
sns.relplot(x = 'Age' , y = 'Niveau de satisfaction', hue = 'Type Contrat', size = 'Salaire base mensuel', 
            style = 'Contrat', data = Dataset_Employes)
```
